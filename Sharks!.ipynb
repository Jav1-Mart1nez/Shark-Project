```python
import re
import pandas as pd

sharks = pd.read_csv("input/attacks.csv", encoding="latin1")
```

```python
"""Como vemos que dentro existen algunos index que tienen espacio entre
el final de la palabra y la comilla, renombramos las columnas para arreglarlo"""

sharks.columns=['Case Number', 'Date', 'Year', 'Type', 'Country', 'Area', 'Location',
       'Activity', 'Name', 'Sex', 'Age', 'Injury', 'Dead', 'Time',
       'Species', 'Investigator or Source', 'pdf', 'href formula', 'href',
       'Case Number.1', 'Case Number.2', 'original order', 'Unnamed: 22',
       'Unnamed: 23']
```

```python
#Comprobamos que los cambios se han hecho efectivos.

sharks.columns
```

```python
"""Una vez consultados los datos, vamos a realizar un análisis para conocer los paises en los que se producen
mayor número de ataques y que procentaje de estos últimos son mortales.Además de esto los agruparemos 
finalmente por décadas, para conocer en que décadas se han producido mayor nº de ataques."""

sharks_attacks=sharks[["Year","Country", "Dead"]]
sharks_attacks
```

```python
#Comprobamos la cantidad de valores "NaN" que tenemos en la columna especies.

sharks_attacks["Year"].isnull().value_counts()
```

```python
sharks_attacks["Country"].isnull().value_counts()
```

```python
sharks_attacks["Dead"].isnull().value_counts()
```

```python
#Eliminamos los valores NaN por orden descendiente de valores nulos, por si así podemos evitarnos algún filtradp.

sharks_attacks1 = sharks_attacks[~sharks_attacks["Dead"].isnull()]
[["Year","Country", "Dead"]]
sharks_attacks1
```

```python
sharks_attacks2 = sharks_attacks1[~sharks_attacks1["Country"].isnull()]
[["Year","Country", "Dead"]]
sharks_attacks2
```

```python
clean_sharks_attacks = sharks_attacks2[~sharks_attacks2["Year"].isnull()]
[["Year","Country", "Dead"]]
clean_sharks_attacks
```

```python
#Eliminamos también los años desconocidos (Valores = 0)para no desvirtuar el análisis.
clean_sharks_attacks = clean_sharks_attacks.drop(clean_sharks_attacks[clean_sharks_attacks['Year']==0.0].index)
clean_sharks_attacks
```
